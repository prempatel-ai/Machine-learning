```python
import matplotlib.pyplot as mp
```

```python
import numpy as np
```

```python
myl=[1,4,5,7,2]
```

```python
mya=np.array(myl)
```

```python
mp.plot(mya)
mp.xlabel('some x axis')

mp.ylabel('some y axis')
mp.show()
```

now for bar we use matploatlib

```python
NL=[1,2,3,4,5,7,3,2,1,2,10,12,15,16,19,10]
```

```python
NA=np.array(NL)
```

```python
mp.hist(NA)
mp.xlabel('some x axis')
mp.ylabel('some y axis')
mp.show()
```

now for scatter plot like dotes in graph

```python
#scatter plot
r=np.array([1,2,4,5,7,8,9])
e=np.array([2,3,4,6,7,8,5])
mp.scatter(r,e)
mp.xlabel('some x axis')
mp.ylabel('some y axis')
mp.show()
```
