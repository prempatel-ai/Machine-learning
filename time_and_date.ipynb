```python
stock['sma_catagory'] = pd.cut(
    stock['sma_50'],
    bins=[0, 50, 100, 150, 200, 250, 300, 350, 400, 450, 500],
    labels=[
        'very_low',
        'low',
        'medium_low',
        'medium',
        'medium_high',
        'high',
        'very_high',
        'extreme_high',
        'ultra_high',
        'peak' 
        
    ],
    right=False  # Optional: excludes the right edge (i.e., [0,50) instead of [0,50])
)
```

```python
stock[['sma_50','sma_catagory']].iloc[[10,11]]
```

```python
stock['sma_catagory'].value_counts(normalize=True)
```

Map:a series method

Substitute each value in a series/feature/colmun with another value , according to:


-A dictionary
-A Function


-- Often used for transforming categorical data

```python
stock.info()
```

```python
stock['day_of_week_name']
```

```python
day_of_week_mapping={
    'Monday':'weekday',
    'Tuesday':'weekday',
    'Thursday':'weekday',
    'Friday':'weekday',
    'sunday':'weekend',
    'saturday':'weekend'
}
stock['day_of_week_name'].map(arg=day_of_week_mapping)
```
